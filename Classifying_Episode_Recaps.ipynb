```python
def rm_punct(txt):
    return ''.join(ch for ch in txt if ch not in punctuation)

def list_of_crimes():
    phrases = [line.lower().strip() for line in
                open('./ref/crimes.txt').read().split('\n')]

    words = list(set([rm_punct(word) for row in phrases
                    for word in row.split() if word not in STOP_WORDS]))
    crimes = phrases + words
    return crimes
```

```python
STOP_WORDS = [word for word in
                open('./ref/stopwords.txt').read().split('\n')]

CRIME_WORDS = list_of_crimes()
```

```python
n_occurances = defaultdict(int)

for word in CRIME_WORDS:
    n = df.corpus.str.count(word).sum()
    n_occurances[word] = n

n_occurances = dict(n_occurances)
n_occurances = sorted(
    n_occurances.iteritems(),
    key=itemgetter(1),
    reverse=True
)
```

```python
n_occurances = [ (crime, freq) for (crime, freq) in n_occurances if freq > 0]
```
